```python
import pandas as pd
import tldextract

# 1) Dataset'i oku
path = "~/Bitirme_Projesi_Dataset_Olusturma/Feature/feature_final_dataset.csv"
df = pd.read_csv(path)

# 2) Domain sütunu yoksa current_url'den üret
if "domain" not in df.columns:
    def extract_domain(url):
        try:
            ext = tldextract.extract(str(url))
            return ext.registered_domain.lower()
        except:
            return ""
    df["domain"] = df["current_url"].apply(extract_domain)

# 3) Domain'i biraz temizle (çok önemli değil ama sağlıklı olur)
df["domain"] = (
    df["domain"]
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace("www.", "", regex=False)
)

# 4) Her domain kaç kez geçmiş? (value_counts)
domain_counts = df["domain"].value_counts()

# 5) En çok tekrar eden ilk 10 domain
top10 = domain_counts.head(10)

print("En çok tekrar eden 10 domain:")
print(top10)
```

```python
import pandas as pd

# ============================
# 1) Feature Dataset Yükleme
# ============================
feature_path = "feature_final_dataset.csv"
feature_df = pd.read_csv(feature_path)

# Domain çıkarma
feature_df["domain"] = (
    feature_df["current_url"]
    .str.extract(r"https?://([^/]+)", expand=False)
    .str.lower()
    .str.strip()
    .str.replace(r"\.$", "", regex=True)
)

print("Feature satır sayısı:", len(feature_df))


# ============================
# 2) WHOIS Dataset Yükleme
# ============================
whois_path = "~/Bitirme_Projesi_Dataset_Olusturma/Feature/whois_feature_merged_clean.csv"
whois_df = pd.read_csv(whois_path)

# Domain normalize
whois_df["domain"] = (
    whois_df["domain"]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"\.$", "", regex=True)
)

# WHOIS duplicate temizliği (eğer kaldıysa)
whois_df = whois_df.drop_duplicates(subset="domain", keep="last")

print("WHOIS satır sayısı:", len(whois_df))


# ============================
# 3) Merge İşlemi
# ============================
merged = feature_df.merge(
    whois_df,
    on="domain",
    how="left"       # Satır sayısını sabit tutar
)

print("Merged satır sayısı:", len(merged))


# ============================
# 4) Final Dataset Kaydet
# ============================
merged.to_csv("feature_final_vselif.csv", index=False)
print("Final dataset kaydedildi: feature_final_vselif.csv")
```

```python
import pandas as pd

# ===============================
# 1) WHOIS dosyasını yükle
# ===============================
path = "~/Bitirme_Projesi_Dataset_Olusturma/Feature/whois_feature_merged.csv"
df = pd.read_csv(path)

print("Orijinal satır sayısı:", len(df))

# ===============================
# 2) Domain kolonundaki boşlukları düzelt
# ===============================
df["domain"] = (
    df["domain"]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"\.$", "", regex=True)
)

# ===============================
# 3) Duplicate domain sayısını kontrol et
# ===============================
dup_count = df["domain"].duplicated().sum()
print("Duplicate domain sayısı:", dup_count)

if dup_count == 0:
    print("\nHiç duplicate yok. Dosya zaten temiz 👍")
else:
    # ===============================
    # 4) Duplicate domainleri sil (son kaydı koru)
    # ===============================
    df_clean = df.drop_duplicates(subset="domain", keep="last")

    removed = len(df) - len(df_clean)

    print(f"\nSilinen duplicate domain sayısı: {removed}")
    print("Temizlenmiş satır sayısı:", len(df_clean))

    # ===============================
    # 5) Temiz WHOIS dosyasını aynı dosyaya yaz (kalıcı)
    # ===============================
    df_clean.to_csv(path, index=False)
    print("\n✔ Duplicate temizliği TAMAMLANDI ve dosya ÜZERİNE yazıldı.")
```
