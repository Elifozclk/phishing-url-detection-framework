```python
import os
import shutil
from pathlib import Path
from datetime import datetime

# === 1️⃣ Ana klasörler ===
OUTPUTS_ROOT = Path(r"C:\Users\elifo\Bitirme_Projesi_Dataset_Olusturma\Makale\outputs")
FINAL_ROOT = Path("FINAL_RESULTS")

# === 2️⃣ Hedef klasör yapısı ===
folder_map = {
    "delta": "01_Global_SHAP/Delta_Comparisons",
    "stability": "01_Global_SHAP/Stability",
    "raw_matrix": "03_Paper_Figures",
    "row_norm": "03_Paper_Figures",
    "top20": "01_Global_SHAP",
    "local": "02_Local_XAI/LocalSHAP",
    "lime": "02_Local_XAI/LIME",
    "diversity": "02_Local_XAI/Diversity_Analysis",
    "compare": "02_Local_XAI/SHAP_vs_LIME",
}

# === 3️⃣ Klasörleri oluştur ===
for sub in set(folder_map.values()):
    (FINAL_ROOT / sub).mkdir(parents=True, exist_ok=True)

# === 4️⃣ Dosya yönlendirme fonksiyonu ===
def classify_destination(file_path):
    name = file_path.name.lower()
    for key, dest in folder_map.items():
        if key in name:
            return FINAL_ROOT / dest / file_path.name
    return None

# === 5️⃣ Çakışma kontrolü (overwrite yapmaz) ===
def safe_copy(src, dst):
    if dst.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        dst = dst.parent / f"{dst.stem}_{timestamp}{dst.suffix}"
    shutil.copy2(src, dst)
    return dst

# === 6️⃣ Recursive tarama ===
copied_files = []
skipped_files = []

for root, dirs, files in os.walk(OUTPUTS_ROOT):
    for file in files:
        src_path = Path(root) / file
        
        # sadece csv, png, json al
        if src_path.suffix.lower() not in [".csv", ".png", ".json"]:
            continue
        
        dst_path = classify_destination(src_path)
        
        if dst_path:
            dst_path.parent.mkdir(parents=True, exist_ok=True)
            final_dst = safe_copy(src_path, dst_path)
            copied_files.append((src_path, final_dst))
        else:
            skipped_files.append(src_path)

print(f"\n✅ Toplam kopyalanan dosya: {len(copied_files)}")
print(f"⚠️ Sınıflandırılamayan dosya: {len(skipped_files)}")

if skipped_files:
    print("\nİlk 10 sınıflandırılamayan dosya:")
    for f in skipped_files[:10]:
        print(" -", f)
```

```python
import os
import shutil
from pathlib import Path
from datetime import datetime

OUTPUTS_ROOT = Path(r"C:\Users\elifo\Bitirme_Projesi_Dataset_Olusturma\Makale\outputs")
FINAL_ROOT = Path("FINAL_RESULTS")

# İstersen cache dosyalarını hiç taşıma (önerilir)
SKIP_CACHE = True

# Dest klasörleri oluştur
dest_folders = [
    "00_Raw_Cache_Skipped",
    "01_Global_SHAP/Per_Model",
    "01_Global_SHAP/Ensemble/Soft",
    "01_Global_SHAP/Ensemble/Hard",
    "01_Global_SHAP/Ensemble/Stacking",
    "01_Global_SHAP/Stability",
    "01_Global_SHAP/Delta_Comparisons",
    "01_Global_SHAP/Feature_Behavior/Dependence",
    "02_Local_XAI/LocalSHAP",
    "02_Local_XAI/LIME",
    "02_Local_XAI/SHAP_vs_LIME",
    "02_Local_XAI/Diversity_Analysis",
    "03_Paper_Figures",
    "04_Tables_For_Thesis",
    "99_Unsorted"
]
for d in dest_folders:
    (FINAL_ROOT / d).mkdir(parents=True, exist_ok=True)

def safe_copy(src: Path, dst: Path) -> Path:
    if dst.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        dst = dst.parent / f"{dst.stem}_{timestamp}{dst.suffix}"
    dst.parent.mkdir(parents=True, exist_ok=True)
    shutil.copy2(src, dst)
    return dst

def get_model_name(path: Path):
    # outputs/<MODEL>/... yapısından model adını çek
    parts = path.parts
    if "outputs" in parts:
        i = parts.index("outputs")
        if i+1 < len(parts):
            return parts[i+1]
    return "UnknownModel"

def classify(src: Path) -> Path:
    p = str(src).lower()

    # ---- 1) Cache skip ----
    if ("cache" in p or "cache_v" in p) and SKIP_CACHE:
        return None  # skip

    # ---- 2) Ensemble özel ----
    if r"outputs\ensembles\ensemblesoft" in p or "ensemblesoft" in p:
        if "stability" in p:
            return FINAL_ROOT / "01_Global_SHAP/Stability" / src.name
        if "lime" in p:
            return FINAL_ROOT / "02_Local_XAI/LIME" / src.name
        if "local_shap" in p or "shap_local" in p:
            return FINAL_ROOT / "02_Local_XAI/LocalSHAP" / src.name
        if "delta" in p:
            return FINAL_ROOT / "01_Global_SHAP/Delta_Comparisons" / src.name
        return FINAL_ROOT / "01_Global_SHAP/Ensemble/Soft" / src.name

    if r"outputs\ensembles\ensemblehard" in p or "ensemblehard" in p:
        if "lime" in p:
            return FINAL_ROOT / "02_Local_XAI/LIME" / src.name
        if "local_shap" in p or "shap_local" in p:
            return FINAL_ROOT / "02_Local_XAI/LocalSHAP" / src.name
        return FINAL_ROOT / "01_Global_SHAP/Ensemble/Hard" / src.name

    if r"outputs\ensembles\ensemblestacking" in p or "ensemblestacking" in p:
        if "lime" in p:
            return FINAL_ROOT / "02_Local_XAI/LIME" / src.name
        if "local_shap" in p or "shap_local" in p:
            return FINAL_ROOT / "02_Local_XAI/LocalSHAP" / src.name
        return FINAL_ROOT / "01_Global_SHAP/Ensemble/Stacking" / src.name

    # ---- 3) Shared/paper/final_package ----
    if r"outputs\_shared" in p:
        if "paper_figures" in p:
            return FINAL_ROOT / "03_Paper_Figures" / src.name
        if "final_package" in p:
            return FINAL_ROOT / "04_Tables_For_Thesis" / src.name
        if "delta_" in src.name.lower():
            return FINAL_ROOT / "01_Global_SHAP/Delta_Comparisons" / src.name
        if "stability" in p:
            return FINAL_ROOT / "01_Global_SHAP/Stability" / src.name
        if "compare" in p or "lime_shap" in p:
            return FINAL_ROOT / "02_Local_XAI/SHAP_vs_LIME" / src.name
        return FINAL_ROOT / "99_Unsorted" / src.name

    # ---- 4) Feature behavior / dependence plots ----
    if "dependence" in p:
        model = get_model_name(src)
        return FINAL_ROOT / "01_Global_SHAP/Feature_Behavior/Dependence" / model / src.name

    # ---- 5) Local XAI per-model ----
    if "shap_local" in p or "local_shap" in p:
        model = get_model_name(src)
        return FINAL_ROOT / "02_Local_XAI/LocalSHAP" / model / src.name

    if "lime_local" in p or ("/lime/" in p.replace("\\", "/") and "explanations" in p):
        model = get_model_name(src)
        return FINAL_ROOT / "02_Local_XAI/LIME" / model / src.name

    # ---- 6) Global SHAP per-model ----
    if r"\global" in p and ("shap" in p):
        model = get_model_name(src)
        return FINAL_ROOT / "01_Global_SHAP/Per_Model" / model / src.name

    # ---- fallback ----
    return FINAL_ROOT / "99_Unsorted" / src.name

# --- Run packing ---
copied = 0
skipped = 0
unsorted = 0
sample_unsorted = []

for root, dirs, files in os.walk(OUTPUTS_ROOT):
    for f in files:
        src = Path(root) / f
        if src.suffix.lower() not in [".csv", ".png", ".json", ".txt"]:
            continue

        dst = classify(src)

        if dst is None:
            skipped += 1
            continue

        final_dst = safe_copy(src, dst)
        copied += 1

        if "99_unsorted" in str(final_dst).lower():
            unsorted += 1
            if len(sample_unsorted) < 10:
                sample_unsorted.append(str(src))

print(f"✅ Kopyalanan dosya: {copied}")
print(f"⏭️ Skip edilen (cache vb): {skipped}")
print(f"⚠️ Unsorted'e düşen: {unsorted}")
if sample_unsorted:
    print("\nUnsorted örnekleri (ilk 10):")
    for s in sample_unsorted:
        print(" -", s)
```
