```python
import pandas as pd

# DoДџru dosya yollarД±
new_url_path = "./Beningdata/new_url.csv"               # рџ”ґ BURASI Г–NEMLД°
success_path = "./Beningdata/success_200_unique.csv"
output_path = "./Beningdata/firstt_url.csv"

# CSV'leri yГјkle
df_new = pd.read_csv(new_url_path)
df_succ = pd.read_csv(success_path)

# BirleЕџtir
df_merged = pd.concat([df_new, df_succ], ignore_index=True)

# Domain / URL'leri normalize et (opsiyonel ama faydalД±)
df_merged["domain"] = df_merged["domain"].astype(str).str.strip().str.lower()
if "main_url" in df_merged.columns:
    df_merged["main_url"] = df_merged["main_url"].astype(str).str.strip()

# Sadece gerГ§ekten URL'i olanlarД± tut (NaN veya boЕџlarД± atmak istiyorsan)
df_merged = df_merged.dropna(subset=["main_url"])
df_merged = df_merged[df_merged["main_url"].str.strip() != ""]

# SADECE domain + main_url bazlД± duplicate temizliДџi
df_merged = df_merged.drop_duplicates(subset=["domain", "main_url"])

# Kaydet
df_merged.to_csv(output_path, index=False)

print("BirleЕџtirme ve duplicate temizliДџi tamamlandД±!")
print(f"Toplam final satД±r sayД±sД±: {len(df_merged)}")
print(f"KayД±t edilen dosya: {output_path}")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import uuid

# GiriЕџ ve Г§Д±kД±Еџ
input_path = "./final_first_step_url.csv"
output_path = "./real_paths_test.csv"

df = pd.read_csv(input_path).head(100)  # Pilot: ilk 100 URL

results = []

def extract_links(base_url):
    """
    GerГ§ek sayfadan link Г§eken fonksiyon
    """
    try:
        r = requests.get(base_url, timeout=10)
        if r.status_code != 200:
            return []

        soup = BeautifulSoup(r.text, "html.parser")
        domain = urlparse(base_url).netloc

        found_paths = []
        for tag in soup.find_all("a", href=True):
            href = tag.get("href")

            # Absolute URLвЂ™ye Г§evir
            full_url = urljoin(base_url, href)

            # AynД± domain mi?
            if urlparse(full_url).netloc == domain:
                path = urlparse(full_url).path
                if path and path not in ["", "/", None]:
                    found_paths.append(path)

        # Duplicate temizle
        found_paths = list(set(found_paths))
        return found_paths

    except Exception:
        return []

for idx, row in df.iterrows():
    base_url = row["main_url"]
    domain = row["domain"]

    print(f"[+] Tarama: {base_url}")

    # Path ID Гјret
    path_id = str(uuid.uuid4())

    paths = extract_links(base_url)

    for p in paths:
        full_url = base_url.rstrip("/") + p
        results.append({
            "domain": domain,
            "main_url": base_url,
            "real_path": p,
            "full_real_url": full_url,
            "path_id": path_id,
            "status": row["status"]
        })

# CSV kaydet
pd.DataFrame(results).to_csv(output_path, index=False)

print("GerГ§ek alt path taramasД± (pilot 100 URL) tamamlandД±.")
print(f"Toplam bulunan gerГ§ek path sayД±sД±: {len(results)}")
print(f"Veriler kaydedildi: {output_path}")
```

```python
import pandas as pd
import time
import uuid
import concurrent.futures
from urllib.parse import urlparse, urljoin

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# ================== KONFД°G ==================

input_path = "./final_first_step_url.csv"   # CSV'ni buraya koy
output_path = "./real_paths_depth2_selenium.csv"

MAX_FIRST_LEVEL_LINKS = 10      # main_url'den toplanacak max link
MAX_SECOND_LEVEL_LINKS = 10     # her first-level URL'den toplanacak max link
MAX_WORKERS = 3                 # aynД± anda kaГ§ domain taransД±n (Selenium olduДџu iГ§in Г§ok yГјksek olmasД±n)

REQUEST_SLEEP = (0.5, 1.2)      # her istek sonrasД± min-max bekleme aralД±ДџД± (saniye)
TEST_LIMIT = 100                # ilk deneme iГ§in: kaГ§ main_url taransД±n? TГјm dataset iГ§in None yapabilirsin.

# Statik dosya uzantД±larД± (indirme linkleri vs. istemiyoruz)
STATIC_EXTENSIONS = (
    ".jpg", ".jpeg", ".png", ".gif", ".svg",
    ".webp", ".ico",
    ".pdf", ".zip", ".rar", ".7z", ".tar", ".gz",
    ".mp4", ".mp3", ".avi", ".mov", ".mkv",
    ".exe", ".msi", ".dmg"
)


# ================== YARDIMCI FONKSД°YONLAR ==================

def create_driver():
    """
    Headless Chrome driver oluЕџturur.
    """
    chrome_options = Options()
    chrome_options.add_argument("--headless=new")   # eski Chrome iГ§in "--headless" da olur
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )

    driver = webdriver.Chrome(options=chrome_options)
    driver.set_page_load_timeout(20)
    return driver


def random_sleep():
    """Her istek sonrasД± kГјГ§Гјk bir bekleme (ban riskini azaltmak iГ§in)."""
    import random
    time.sleep(random.uniform(*REQUEST_SLEEP))


def clean_and_filter_link(base_url, href, allowed_domain):
    """
    href'i absolute URL'ye Г§evirir, filtreler ve (full_url, path) dГ¶ner.
    Д°stemediДџimiz Еџeyleri eliyoruz:
      - mailto:, javascript:, tel:
      - dД±Еџ domain
      - boЕџ path veya '/'
      - statik dosyalar (jpg, pdf vs)
    """
    if not href:
        return None

    href = href.strip()

    # Д°stemediДџimiz protokoller
    if href.startswith(("mailto:", "javascript:", "tel:", "#")):
        return None

    full = urljoin(base_url, href)
    parsed = urlparse(full)

    # Sadece http/https
    if parsed.scheme not in ("http", "https"):
        return None

    # AynД± domain deДџilse alma
    if parsed.netloc != allowed_domain:
        return None

    path = parsed.path or "/"

    # Ana sayfa path'i istemiyorsak:
    if path in ("", "/"):
        return None

    # Statik dosyalarД± ele
    for ext in STATIC_EXTENSIONS:
        if path.lower().endswith(ext):
            return None

    return full, path


def extract_links_js(driver, base_url, domain, max_links):
    """
    Selenium ile JS-render edilmiЕџ sayfadan link toplar.
    Sadece aynД± domain ve filtreleri geГ§en linkleri dГ¶ner.
    max_links kadar benzersiz path dГ¶ndГјrГјr.
    """
    links = []
    seen_paths = set()

    try:
        driver.get(base_url)
        # JS'in Г§alД±ЕџmasД± iГ§in biraz bekle
        time.sleep(1.5)

        anchors = driver.find_elements(By.TAG_NAME, "a")

        for a in anchors:
            href = a.get_attribute("href")
            # BazД± sitelerde data-href vs. kullanД±labilir, istersen buraya ekleyebiliriz.
            if not href:
                continue

            cleaned = clean_and_filter_link(base_url, href, domain)
            if not cleaned:
                continue

            full_url, path = cleaned

            if path in seen_paths:
                continue

            seen_paths.add(path)
            links.append((full_url, path))

            if len(links) >= max_links:
                break

    except Exception as e:
        # Hata olursa o sayfayД± pas geГ§
        # print(f"[!] extract_links_js hata: {base_url} -> {e}")
        return []

    random_sleep()
    return links


# ================== ASIL CRAWLER (Depth=2) ==================

def crawl_main_url(row):
    """
    Tek bir main_url satД±rД± iГ§in:
      - Level 1: main_url'den MAX_FIRST_LEVEL_LINKS kadar link
      - Level 2: her Level 1 linkinden MAX_SECOND_LEVEL_LINKS kadar link
    SonuГ§ta satД±r listesi (dict) dГ¶ner.
    Her thread kendi Selenium driver'Д±nД± kullanД±r.
    """
    domain = row["domain"]
    main_url = row["main_url"]
    status = row.get("status", None)

    # Domain netloc doДџrulama iГ§in
    parsed_main = urlparse(main_url)
    base_domain = parsed_main.netloc
    if not base_domain:
        # GeГ§ersiz URL
        return []

    path_tree_id = str(uuid.uuid4())
    results = []
    visited = set()  # aynД± current_url'e tekrar gitmemek iГ§in

    driver = None
    try:
        driver = create_driver()

        # ---- LEVEL 1 ----
        first_level_links = extract_links_js(
            driver=driver,
            base_url=main_url,
            domain=base_domain,
            max_links=MAX_FIRST_LEVEL_LINKS
        )

        for full_url, path in first_level_links:
            if full_url in visited:
                continue
            visited.add(full_url)

            results.append({
                "domain": domain,
                "main_url": main_url,
                "parent_url": main_url,
                "current_url": full_url,
                "real_path": path,
                "depth": 1,
                "status": status,
                "path_tree_id": path_tree_id
            })

        # ---- LEVEL 2 ----
        for full_url_level1, path_level1 in first_level_links:
            second_level_links = extract_links_js(
                driver=driver,
                base_url=full_url_level1,
                domain=base_domain,
                max_links=MAX_SECOND_LEVEL_LINKS
            )

            for full_url2, path2 in second_level_links:
                if full_url2 in visited:
                    continue
                visited.add(full_url2)

                results.append({
                    "domain": domain,
                    "main_url": main_url,
                    "parent_url": full_url_level1,
                    "current_url": full_url2,
                    "real_path": path2,
                    "depth": 2,
                    "status": status,
                    "path_tree_id": path_tree_id
                })

    except Exception as e:
        # print(f"[!] crawl_main_url hata: {main_url} -> {e}")
        pass
    finally:
        if driver is not None:
            try:
                driver.quit()
            except Exception:
                pass

    return results


# ================== ANA AKIЕћ ==================

# CSV'yi yГјkle
df = pd.read_csv(input_path)

# Д°lk test iГ§in subset al; her Еџey yolundaysa bu satД±rД± kaldД±rabilirsin
if TEST_LIMIT is not None:
    df = df.head(TEST_LIMIT)

print(f"Toplam main_url sayД±sД± (bu koЕџuda): {len(df)}")

all_results = []

with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    future_to_idx = {
        executor.submit(crawl_main_url, row): idx
        for idx, row in df.iterrows()
    }

    for future in concurrent.futures.as_completed(future_to_idx):
        idx = future_to_idx[future]
        try:
            res = future.result()
            if res:
                all_results.extend(res)
        except Exception as e:
            print(f"[!] Thread hatasД±, satД±r {idx}: {e}")

# SonuГ§larД± CSV'ye yaz
if all_results:
    out_df = pd.DataFrame(all_results)
    out_df.to_csv(output_path, index=False)
    print(f"Toplam bulunan URL sayД±sД±: {len(out_df)}")
    print(f"Kaydedilen dosya: {output_path}")
else:
    print("HiГ§ URL bulunamadД± veya tГјm istekler hata verdi.")
```

```python
import pandas as pd
import time
import uuid
import concurrent.futures
import os
from urllib.parse import urlparse, urljoin
from datetime import datetime

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# ================== KONFД°G ==================

input_path = "./final_first_step_url.csv"     # Girdi CSV
output_path = "./real_paths_depth2_vs2_selenium.csv"      # URLвЂ™lerin kaydedileceДџi dosya
no_url_output_path = "./no_urls_depth2_selenium.csv"  # HiГ§ URL Гјretemeyen main_urlвЂ™ler

MAX_FIRST_LEVEL_LINKS = 5       # Depth 1: main_url'den toplanacak max link sayД±sД±
MAX_SECOND_LEVEL_LINKS = 10     # Depth 2: her first-level URL'den toplanacak max link sayД±sД±

MAX_WORKERS = 3                 # AynД± anda kaГ§ domain taransД±n (Selenium, Г§ok abartma)
REQUEST_SLEEP = (0.5, 1.2)      # Her istek sonrasД± min-max bekleme (saniye)

CHECKPOINT_EVERY = 20           # Her 20 main_url'de bir ara kayД±t
TEST_LIMIT = 100                # Д°lk 100 main_url ile test

# Statik dosya uzantД±larД± (indirme linkleri vs. istemiyoruz)
STATIC_EXTENSIONS = (
    ".jpg", ".jpeg", ".png", ".gif", ".svg",
    ".webp", ".ico",
    ".pdf", ".zip", ".rar", ".7z", ".tar", ".gz",
    ".mp4", ".mp3", ".avi", ".mov", ".mkv",
    ".exe", ".msi", ".dmg"
)


# ================== YARDIMCI FONKSД°YONLAR ==================

def create_driver():
    """
    Headless Chrome driver oluЕџturur.
    """
    chrome_options = Options()
    chrome_options.add_argument("--headless=new")   # Gerekirse "--headless" yap
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )

    driver = webdriver.Chrome(options=chrome_options)
    driver.set_page_load_timeout(20)
    return driver


def random_sleep():
    """Her istek sonrasД± kГјГ§Гјk bir bekleme (ban riskini azaltmak iГ§in)."""
    import random
    time.sleep(random.uniform(*REQUEST_SLEEP))


def clean_and_filter_link(base_url, href, allowed_domain):
    """
    href'i absolute URL'ye Г§evirir, filtreler ve (full_url, path) dГ¶ner.
    Д°stemediДџimiz Еџeyleri eliyoruz:
      - mailto:, javascript:, tel:, #
      - dД±Еџ domain
      - boЕџ path veya '/'
      - statik dosyalar (jpg, pdf vs)
    """
    if not href:
        return None

    href = href.strip()

    # Д°stemediДџimiz protokoller
    if href.startswith(("mailto:", "javascript:", "tel:", "#")):
        return None

    full = urljoin(base_url, href)
    parsed = urlparse(full)

    # Sadece http/https
    if parsed.scheme not in ("http", "https"):
        return None

    # AynД± domain deДџilse alma
    if parsed.netloc != allowed_domain:
        return None

    path = parsed.path or "/"

    # Ana sayfa path'i istemiyorsak:
    if path in ("", "/"):
        return None

    # Statik dosyalarД± ele
    for ext in STATIC_EXTENSIONS:
        if path.lower().endswith(ext):
            return None

    return full, path


def extract_links_js(driver, base_url, domain, max_links):
    """
    Selenium ile JS-render edilmiЕџ sayfadan link toplar.
    Sadece aynД± domain ve filtreleri geГ§en linkleri dГ¶ner.
    max_links kadar benzersiz path dГ¶ndГјrГјr.
    """
    links = []
    seen_paths = set()

    try:
        driver.get(base_url)
        # JS'in Г§alД±ЕџmasД± iГ§in biraz bekle
        time.sleep(1.5)

        anchors = driver.find_elements(By.TAG_NAME, "a")

        for a in anchors:
            href = a.get_attribute("href")
            if not href:
                continue

            cleaned = clean_and_filter_link(base_url, href, domain)
            if not cleaned:
                continue

            full_url, path = cleaned

            if path in seen_paths:
                continue

            seen_paths.add(path)
            links.append((full_url, path))

            if len(links) >= max_links:
                break

    except Exception:
        # Hata olursa o sayfayД± pas geГ§
        return []

    random_sleep()
    return links


# ================== ASIL CRAWLER (Depth=2) ==================

def crawl_main_url(row):
    """
    Tek bir main_url satД±rД± iГ§in:
      - Level 1: main_url'den MAX_FIRST_LEVEL_LINKS kadar link
      - Level 2: her Level 1 linkinden MAX_SECOND_LEVEL_LINKS kadar link
    SonuГ§ta satД±r listesi (dict) dГ¶ner.
    """
    domain = row["domain"]
    main_url = row["main_url"]
    status = row.get("status", None)

    parsed_main = urlparse(main_url)
    base_domain = parsed_main.netloc
    if not base_domain:
        # GeГ§ersiz URL
        return []

    path_tree_id = str(uuid.uuid4())
    results = []
    visited = set()  # aynД± current_url'e tekrar gitmemek iГ§in

    driver = None
    try:
        driver = create_driver()

        # ---- LEVEL 1 (Depth = 1) ----
        first_level_links = extract_links_js(
            driver=driver,
            base_url=main_url,
            domain=base_domain,
            max_links=MAX_FIRST_LEVEL_LINKS
        )

        for full_url, path in first_level_links:
            if full_url in visited:
                continue
            visited.add(full_url)

            results.append({
                "domain": domain,
                "main_url": main_url,
                "parent_url": main_url,
                "current_url": full_url,
                "real_path": path,
                "depth": 1,
                "status": status,
                "path_tree_id": path_tree_id
            })

        # ---- LEVEL 2 (Depth = 2) ----
        for full_url_level1, path_level1 in first_level_links:
            second_level_links = extract_links_js(
                driver=driver,
                base_url=full_url_level1,
                domain=base_domain,
                max_links=MAX_SECOND_LEVEL_LINKS
            )

            for full_url2, path2 in second_level_links:
                if full_url2 in visited:
                    continue
                visited.add(full_url2)

                results.append({
                    "domain": domain,
                    "main_url": main_url,
                    "parent_url": full_url_level1,
                    "current_url": full_url2,
                    "real_path": path2,
                    "depth": 2,
                    "status": status,
                    "path_tree_id": path_tree_id
                })

    except Exception:
        pass
    finally:
        if driver is not None:
            try:
                driver.quit()
            except Exception:
                pass

    return results


# ================== ANA AKIЕћ (CHECKPOINT'LД°) ==================

# Girdi CSV'yi yГјkle
df = pd.read_csv(input_path)

# Д°lk 100lГјk test
if TEST_LIMIT is not None:
    df = df.head(TEST_LIMIT)

print(f"Toplam main_url sayД±sД± (bu koЕџuda): {len(df)}")

# Eski output dosyalarД±nД± tekrar kullanmak istemiyorsan, manuel silmen iyi olur:
# os.remove(output_path)  # dikkat!
# os.remove(no_url_output_path)

buffer_results = []      # URL kayД±tlarД± iГ§in buffer
no_url_buffer = []       # URL Гјretemeyen main_url'ler iГ§in buffer

processed_main_urls = 0
total_urls_written = 0
total_no_url_domains = 0

def flush_buffers():
    """Ara kayД±t: buffer'daki verileri CSV'lere append eder ve ekrana Г¶zet basar."""
    global buffer_results, no_url_buffer, total_urls_written, total_no_url_domains

    # URL'ler iГ§in
    if buffer_results:
        df_out = pd.DataFrame(buffer_results)
        file_exists = os.path.exists(output_path)
        df_out.to_csv(
            output_path,
            mode='a',
            index=False,
            header=not file_exists  # dosya yoksa header yaz
        )
        total_urls_written += len(df_out)
        buffer_results = []

    # URL Гјretemeyen domainler iГ§in
    if no_url_buffer:
        df_fail = pd.DataFrame(no_url_buffer)
        file_exists_fail = os.path.exists(no_url_output_path)
        df_fail.to_csv(
            no_url_output_path,
            mode='a',
            index=False,
            header=not file_exists_fail
        )
        total_no_url_domains += len(df_fail)
        no_url_buffer = []

    # Konsola bilgi
    print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}] "
          f"Ећu ana kadar iЕџlenen main_url sayД±sД±: {processed_main_urls} | "
          f"Toplam yazД±lan URL sayД±sД±: {total_urls_written} | "
          f"HiГ§ URL Гјretemeyen domain sayД±sД±: {total_no_url_domains}")


with concurrent.futures.ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
    future_to_meta = {}

    for idx, row in df.iterrows():
        future = executor.submit(crawl_main_url, row)
        future_to_meta[future] = {
            "idx": idx,
            "domain": row["domain"],
            "main_url": row["main_url"],
            "status": row.get("status", None)
        }

    for future in concurrent.futures.as_completed(future_to_meta):
        meta = future_to_meta[future]
        try:
            res = future.result()
        except Exception as e:
            print(f"[!] Thread hatasД±, satД±r {meta['idx']}: {e}")
            res = []

        processed_main_urls += 1

        if res:
            # URL Гјretenler
            buffer_results.extend(res)
        else:
            # HiГ§ URL Гјretemeyen main_url
            no_url_buffer.append({
                "domain": meta["domain"],
                "main_url": meta["main_url"],
                "status": meta["status"]
            })

        # Her CHECKPOINT_EVERY main_url'de bir ara kayД±t
        if processed_main_urls % CHECKPOINT_EVERY == 0:
            flush_buffers()

# DГ¶ngГј bittiДџinde elde kalan buffer'larД± da yaz
flush_buffers()

print("Tarama tamamlandД±.")
print(f"Toplam yazД±lan URL sayД±sД±: {total_urls_written}")
print(f"Toplam hiГ§ URL Гјretemeyen main_url sayД±sД±: {total_no_url_domains}")
print(f"URL'ler CSV: {output_path}")
print(f"URL Гјretemeyenler CSV: {no_url_output_path}")
```

```python
import pandas as pd
import random

# CSV yГјkle
df = pd.read_csv("real_paths_normal_playwright_final.csv")

# Depth=2 olanlar
depth2 = df[df["depth"] == 2]

# Depth=2 sayД±sД± > 9 olan main_urlвЂ™ler
depth2_counts = depth2.groupby("main_url")["current_url"].count().reset_index()
valid_main_urls = depth2_counts[depth2_counts["current_url"] >= 8]["main_url"].tolist()

print("Depth=2 > 9 olan main_url sayД±sД±:", len(valid_main_urls))

final_selected_rows = []

for main_url in valid_main_urls:
    sub = depth2[depth2["main_url"] == main_url]

    # Parent'e gГ¶re grupla
    parent_groups = {p: g for p, g in sub.groupby("parent_url")}

    selected = []

    # 1) Her parent_urlвЂ™den 1 tane al в†’ Г§eЕџitlilik maksimize
    for parent, g in parent_groups.items():
        selected.append(g.sample(1))
        if len(selected) == 8:
            break

    # 2) EДџer parent_url sayД±sД± < 8 ise, rastgele tamamla
    if len(selected) < 8:
        remaining_needed = 8 - len(selected)

        # daha Г¶nce seГ§ilmeyen depth2 pathlerden rastgele seГ§
        already_indices = pd.concat(selected).index
        pool = sub[~sub.index.isin(already_indices)]

        if len(pool) >= remaining_needed:
            extra = pool.sample(remaining_needed)
            selected.append(extra)
        else:
            selected.append(pool)

    # SonuГ§larД± ekle
    final_selected_rows.append(pd.concat(selected))

# Nihai dataframe
final_df = pd.concat(final_selected_rows).reset_index(drop=True)

print("Toplam seГ§ilen path sayД±sД±:", len(final_df))

# CSV olarak kaydet
output_csv = "selected_diverse_depth2_paths.csv"
final_df.to_csv(output_csv, index=False)

print(f"\nCSV baЕџarД±yla kaydedildi в†’ {output_csv}")
```

```python
import pandas as pd
import random

df = pd.read_csv("real_paths_normal_playwright_final.csv")

OUTPUT_SIZE = 8  # hedef path sayД±sД±
final_rows = []

# Her main_url iГ§in
main_urls = df["main_url"].unique()

for main_url in main_urls:
    
    d1 = df[(df["main_url"] == main_url) & (df["depth"] == 1)]
    d2 = df[(df["main_url"] == main_url) & (df["depth"] == 2)]
    
    total_paths = len(d1) + len(d2)
    
    # ==========================================================
    # DURUM 1 вЂ” Toplam path 8вЂ™den az в†’ Hepsini direkt al
    # ==========================================================
    if total_paths <= OUTPUT_SIZE:
        if total_paths > 0:
            final_rows.append(pd.concat([d1, d2]))
        continue  # bu main_url iГ§in daha seГ§im yapД±lmayacak
    
    # ==========================================================
    # DURUM 2 вЂ” Toplam path 8вЂ™den fazla в†’ AkД±llД± seГ§ime geГ§
    # ==========================================================
    
    selected = []
    
    # ---------- AЕћAMA 1: Parent Г§eЕџitliliДџini maksimize (depth 2) ----------
    parent_groups = {p: g for p, g in d2.groupby("parent_url")}
    
    for parent, group in parent_groups.items():
        selected.append(group.sample(1))  # her parentвЂ™tan 1 child
        if len(selected) == OUTPUT_SIZE:
            break
    
    # ---------- AЕћAMA 2: Depth2 havuzundan rastgele ekle ----------
    if len(selected) < OUTPUT_SIZE and len(d2) > len(selected):
        
        already_idx = pd.concat(selected).index if selected else []
        remaining_d2 = d2[~d2.index.isin(already_idx)]
        
        need = OUTPUT_SIZE - len(selected)
        if len(remaining_d2) > 0:
            selected.append(remaining_d2.sample(min(len(remaining_d2), need)))
    
    # ---------- AЕћAMA 3: Depth1 ile tamamla ----------
    if len(selected) < OUTPUT_SIZE:
        need = OUTPUT_SIZE - len(selected)
        if len(d1) > 0:
            selected.append(d1.sample(min(len(d1), need)))
    
    # Son seГ§ilenleri birleЕџtir
    if selected:
        final_rows.append(pd.concat(selected))


# Final output
final_df = pd.concat(final_rows).reset_index(drop=True)

print("Toplam seГ§ilen path:", len(final_df))

output_path = "smart_selected_paths.csv"
final_df.to_csv(output_path, index=False)

print(f"Kaydedildi в†’ {output_path}")
```

```python
import pandas as pd
import numpy as np
import random

df = pd.read_csv("real_paths_normal_playwright_final.csv")
df["depth"] = pd.to_numeric(df["depth"], errors="coerce").fillna(0).astype(int)

print("Toplam satД±r:", len(df))
print("Toplam main_url:", df["main_url"].nunique())
```

```python
groups = {m: g for m, g in df.groupby("main_url")}
print("HazД±r grup sayД±sД±:", len(groups))
```

```python
OUTPUT_SIZE = 8

def fast_select(group):
    d1 = group[group["depth"] == 1]
    d2 = group[group["depth"] == 2]

    total = len(d1) + len(d2)

    # EДџer toplam path <= 8 ise direkt hepsini al
    if total <= OUTPUT_SIZE:
        return group

    selected = []

    # AЕџama 1 вЂ” parent Г§eЕџitliliДџini maksimize et
    parent_groups = list(d2.groupby("parent_url"))
    random.shuffle(parent_groups)

    for parent, grp in parent_groups:
        selected.append(grp.sample(1))
        if len(selected) == OUTPUT_SIZE:
            return pd.concat(selected)

    # AЕџama 2 вЂ” depth2вЂ™den random doldur
    if len(selected) < OUTPUT_SIZE and len(d2) > 0:
        needed = OUTPUT_SIZE - len(selected)
        remaining = d2.loc[~d2.index.isin(pd.concat(selected).index)]
        if len(remaining) > 0:
            selected.append(
                remaining.sample(min(needed, len(remaining)))
            )

    # AЕџama 3 вЂ” depth1вЂ™den tamamla
    combined = pd.concat(selected) if selected else pd.DataFrame()

    if len(combined) < OUTPUT_SIZE and len(d1) > 0:
        needed = OUTPUT_SIZE - len(combined)
        selected_d1 = d1.sample(min(needed, len(d1)))
        combined = pd.concat([combined, selected_d1])

    return combined
```

```python
final_list = []

for main_url, group in groups.items():
    final_list.append(fast_select(group))

final_df = pd.concat(final_list).reset_index(drop=True)

print("Toplam seГ§ilen path:", len(final_df))
```
