```python
import os
import pandas as pd
from datetime import datetime, timedelta, timezone
from urllib.parse import urlparse
import re

# ---- Klasör ve dosya yolu ----
base_folder = r"C:\Users\elifo\Bitirme_Projesi_Dataset_Olusturma\Phishingdata_phishtank"
pt_csv = os.path.join(base_folder, "online-valid.csv")

# ---- 1️⃣ CSV'yi oku (önizleme) ----
print("📄 Okunuyor:", pt_csv)
tmp = pd.read_csv(pt_csv, dtype=str, nrows=5, low_memory=False)
print("Kolonlar:", tmp.columns.tolist())

# ---- 2️⃣ Tarih sütununu otomatik bul ----
possible_date_cols = [c for c in tmp.columns if any(x in c.lower() for x in ["submission", "verified", "date", "time"])]
date_col = possible_date_cols[0] if possible_date_cols else None
print("⏱️ Tarih kolonu:", date_col)

# ---- 3️⃣ Tüm dosyayı parçalı oku ve son 1 yıl filtrele ----
cutoff = datetime.now(timezone.utc) - timedelta(days=365)
chunksize = 200_000
kept_chunks = []
total_in = total_kept = 0

for chunk in pd.read_csv(pt_csv, dtype=str, chunksize=chunksize, low_memory=True):
    total_in += len(chunk)
    if date_col:
        chunk[date_col + "_parsed"] = pd.to_datetime(chunk[date_col], errors="coerce", utc=True)
        sel = chunk[chunk[date_col + "_parsed"] >= cutoff]
    else:
        sel = chunk.copy()
    if not sel.empty:
        kept_chunks.append(sel)
        total_kept += len(sel)

print(f"Toplam {total_in} satır okundu; son 1 yılda {total_kept} kayıt bulundu.")

if kept_chunks:
    df_pt_recent = pd.concat(kept_chunks, ignore_index=True)
else:
    df_pt_recent = pd.DataFrame()

# ---- 4️⃣ URL sütununu tespit et ----
if not df_pt_recent.empty:
    possible_url_cols = [c for c in df_pt_recent.columns if any(k in c.lower() for k in ["url", "link", "uri"])]
    url_col = possible_url_cols[0] if possible_url_cols else df_pt_recent.columns[0]
    print("🔗 URL kolonu:", url_col)
else:
    print("⚠️ Hiç kayıt bulunamadı, URL tespiti atlandı.")
    url_col = None

# ---- 5️⃣ URL normalize + dedupe ----
def normalize_url(u):
    if pd.isna(u): 
        return u
    u = str(u).strip()
    u = re.sub(r"^https?://", "", u, flags=re.I)
    u = u.rstrip("/")
    try:
        p = urlparse("http://" + u)
        host = p.netloc.lower()
        if host.startswith("www."):
            host = host[4:]
        path = p.path.rstrip("/")
        qs = ("?" + p.query) if p.query else ""
        return host + path + qs
    except:
        return u.lower()

if url_col and not df_pt_recent.empty:
    df_pt_recent["url_norm"] = df_pt_recent[url_col].astype(str).apply(normalize_url)
    before = len(df_pt_recent)
    df_pt_recent = df_pt_recent.drop_duplicates(subset=["url_norm"])
    after = len(df_pt_recent)
    print(f"🔄 Tekrarlar kaldırıldı: {before} → {after}")

# ---- 6️⃣ Kaydet ----
out_csv = os.path.join(base_folder, "Phishingdata_phishtank_last1y.csv")
df_pt_recent.to_csv(out_csv, index=False)
print(f"✅ Kaydedildi: {out_csv}")
print(f"Satır sayısı: {len(df_pt_recent)}")
```

```python
import os
import pandas as pd

# ---- 1️⃣ Dosya yolları ----
base_folder = r"C:\Users\elifo\Bitirme_Projesi_Dataset_Olusturma"
openphish_path = os.path.join(base_folder, "Phishingdata_openphish", "Phishingdata_openphish_last1y.csv")
phishtank_path = os.path.join(base_folder, "Phishingdata_phishtank", "Phishingdata_phishtank_last1y.csv")
combined_path = os.path.join(base_folder, "Phishingdata_all_phishdata.csv")

# ---- 2️⃣ Dosyaları oku (yalnızca url kolonları) ----
def load_urls(path):
    if not os.path.exists(path):
        print(f"⚠️ Dosya bulunamadı: {path}")
        return pd.DataFrame(columns=["url"])
    df = pd.read_csv(path, dtype=str, usecols=lambda c: "url" in c.lower(), low_memory=False)
    print(f"📄 {os.path.basename(path)} yüklendi. Satır sayısı: {len(df)}")
    # Kolon ismini normalize et
    url_col = [c for c in df.columns if "url" in c.lower()][0]
    df = df.rename(columns={url_col: "url"})
    return df[["url"]]

df_open = load_urls(openphish_path)
df_pt = load_urls(phishtank_path)

# ---- 3️⃣ Birleştir ve tekrarlanan URL’leri kaldır ----
df_all = pd.concat([df_open, df_pt], ignore_index=True)
print(f"\n🔗 Toplam URL sayısı (birleşik): {len(df_all)}")

df_all.drop_duplicates(subset=["url"], inplace=True)
print(f"✅ Tekrarlardan arındırıldı: {len(df_all)} benzersiz URL kaldı.")

# ---- 4️⃣ Kaydet ----
df_all.to_csv(combined_path, index=False)
print(f"💾 Kaydedildi: {combined_path}")
```

```python
import pandas as pd
import re
from urllib.parse import urlparse

# Dosya yolu
path = r"C:\Users\elifo\Bitirme_Projesi_Dataset_Olusturma\Phishingdata_all_phishdata.csv"

# 1️⃣ Dosyayı oku
df = pd.read_csv(path, dtype=str)
print("📄 Orijinal satır sayısı:", len(df))

# 2️⃣ URL normalize fonksiyonu
def normalize_url(u):
    if pd.isna(u):
        return u
    u = str(u).strip()
    # Protokolü kaldır (http / https)
    u = re.sub(r"^https?://", "", u, flags=re.I)
    # Sondaki '/' karakterini kaldır
    u = u.rstrip('/')
    try:
        p = urlparse("http://" + u)
        host = p.netloc.lower()
        if host.startswith("www."):
            host = host[4:]
        path = p.path.rstrip('/')
        qs = ("?" + p.query) if p.query else ""
        return host + path + qs
    except:
        return u.lower()

# 3️⃣ Normalizasyon ve tekrarların kaldırılması
df["url_norm"] = df["url"].apply(normalize_url)
before = len(df)
df_clean = df.drop_duplicates(subset=["url_norm"]).copy()
after = len(df_clean)

print(f"🔄 Tekrarlar kaldırıldı: {before} → {after}")

# 4️⃣ Artık normalize sütuna gerek yoksa silebiliriz
df_clean = df_clean.drop(columns=["url_norm"])

# 5️⃣ Temiz dosyayı kaydet
clean_path = path.replace(".csv", "_clean.csv")
df_clean.to_csv(clean_path, index=False)
print(f"✅ Temiz dosya kaydedildi: {clean_path}")
```
